```python
def returnSum(dict): 
     sum = 0
     for i in dict.values(): 
           sum = sum + i 
     return sum

def returnMinus(dict): 
     min = 0
     for i in dict.values(): 
           min = min - i 
     return min
  
dict = {'a': 1, 'b':2, 'c':3} 
print("Sum :", returnSum(dict)) 
print("Minus :", returnMinus(dict)) 

print(dict['a'] + dict['b'] * dict['c'])
```
